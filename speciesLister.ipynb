```python
omit_list = [] # Build list of known species to omit

with open("results/lists/omit_test_lenient.txt", "r") as myfile:
    for line in myfile:
        omit_list.append(line.rstrip()[:line.index("_GC")])

omit_list
```

```python
samnCSVpath = 'metadata_subsets_filt/'
os.listdir(samnCSVpath)

sp_list_raw = []
sp_list = []
for f in os.listdir(samnCSVpath):
    df = pd.read_csv(samnCSVpath + f, header=0, sep=",")
    sp_list_raw.append(df["Organism"].to_list()[0].replace(" ", "_"))

with open("sp_list.txt", "w") as myfile:
    for sp in sp_list_raw:
        if sp not in omit_list:
            sp_list.append(sp)
            myfile.write(sp + "\n")
```

```python
# True if something in sp_list doesn't belong
any([True if i in omit_list else False for i in sp_list])
```

# Mammals List and Generation Lengths


```python
import os
import pandas as pd
import numpy as np
```

```python
path = "drive/MyDrive/MSMC-Curve-Analysis/"

# Retrieve names of our curves
mammals_data_names = os.listdir(path + "msmc_curve_data/mammals_part_1")

# Retrieve generation length data from xls
mammal_analysis_path = path + "MSMC-Exploratory-Analysis/mammal-analysis/"
mammal_gls_df = pd.read_excel(mammal_analysis_path + "mammal_gls.xls")
gls_mammals_df = mammal_gls_df[["Scientific_name", "GenerationLength_d"]]
gls_mammals_df["Scientific_name"] = [name.replace(" ", "_") for name in gls_mammals_df["Scientific_name"].tolist()]
gls_mammals_df["GenerationLength_d"] = gls_mammals_df["GenerationLength_d"]/365
# Make new df conform to our standard format found in drive/MyDrive/MSMC-Curve-Analysis/generation_lengths/
gls_mammals_df = gls_mammals_df.rename(columns={'Scientific_name': 'taxa', 'GenerationLength_d': 'GenLength'})
gls_mammals_df.index = gls_mammals_df["taxa"]
gls_mammals_df = gls_mammals_df.drop(["taxa"], axis=1)
# Save file
gls_mammals_df.to_csv(path + "generation_lengths/mammal_gls.txt", sep = "\t")
```
